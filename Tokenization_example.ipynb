```python
corpus="""Hello Welcome, to Kailasa the abode of Shiva.
Jai Shiv Shankar! Har har shankar.
"""
```

```python
print(corpus)
```

```python
#Tokenization
#Sentence --> Paragraph
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize
```

```python
documents=sent_tokenize(corpus)
```

```python
documents
```

```python
#Tokenization
#Paragraph into words
#Senetence into words

from nltk.tokenize import word_tokenize

word_tokenize(corpus)
```

```python
for sentence in documents:
  print(word_tokenize(sentence))
```

```python
from nltk import wordpunct_tokenize
#punctuations will be considered as seperate word here 3
len(wordpunct_tokenize(corpus))
```

```python
from nltk.tokenize import TreebankWordTokenizer

#full stop will not be treated as seprate word

tokenizer = TreebankWordTokenizer()

tokenizer.tokenize(corpus)
```
